```python

for num,f in enumerate(os.listdir(fpath)):
    file=os.path.join(fpath,f)
    ans=[]
    with open(file,'r') as ff:
        while True:
            lines = ff.readline()
            if not lines:
                break
                pass
            n,c=lines.split()
            ans.append(c)
    for i in range(8):
        if (4*i == 8 or 4*i==20):
            if ans[4*i]=='1':
                v+=1
            else:
                pv+=1
        if 4*i+1==5 or 4*i+1==17:
            if ans[4*i+1]=='1':
                pb+=1
            else: 
                b+=1
        if 4*i+2==14 or 4*i+2==26:
            if ans[4*i+2]=='1':
                e+=1
            else: 
                pe+=1
        if 4*i+3==3 or 4*i+3==31:
            if ans[4*i+3]=='1':
                pk+=1
            else: 
                k+=1
        if (4*i == 12 or 4*i==20):
            if ans[32+4*i]=='1':
                av+=1
            else: 
                apv+=1
        if 4*i+1==9 or 4*i+1==17:
            if ans[32+4*i+1]=='1':
                apb+=1
            else: 
                ab+=1
        if 4*i+2==2 or 4*i+2==30:
            if ans[32+4*i+2]=='1':
                ae+=1
            else: 
                ape+=1
        if 4*i+3==7 or 4*i+3==27:
            if ans[32+4*i+3]=='1':
                apk+=1
            else: 
                ak+=1
```

```python
def best_bag(w,v,b):
    n=len(w)
    dp=[[0]*(b+1) for _ in range(n+1)]
    w.insert(0,0)
    v.insert(0,0)
    for i in range(1,n+1):
        for j in range(1,b+1):
            if w[i]<=j:
                dp[i][j]=max(dp[i-1][j],dp[i-1][j-w[i]]+v[i])
            else:
                dp[i][j]=dp[i-1][j]
    print(dp[n][b])
```

```python

def book(start, end):
    """
    :type start: int
    :type end: int
    :rtype: bool
    """
    if len(booked)==0:
        booked.append([start,end])
        return True
    for i in booked:
        
        if start >= i[0] and start < i[1]:
            return False
        if end > i[0] and end <= i[1]:
            return False
    booked.append([start,end])
    return True
```
